```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Set the display option to show all columns and rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
cookie = pd.read_csv('../data/cookie.csv')
cracker = pd.read_csv('../data/crackers.csv')
cream = pd.read_csv('../data/cream.csv')
wafer = pd.read_csv('../data/wafer.csv')
```

```python
cookie.head()
```

```python
cookie = cookie[['Type','Names','Per Serving', 'Total Fat', 'Sodium', 'Sugars']]
cookie.rename(columns={
    'Type':'type',
    'Names': 'product',
    'Per Serving': 'per_serving',
    'Total Fat': 'total_fat',
    'Sodium': 'sodium',
    'Sugars': 'sugars'
},inplace=True)
cookie.head()
```

```python
cracker.head()
```

```python
cracker = cracker[['Type','product','Per Serving','Total Fat', 'Sodium', 'Sugars']]
cracker.rename(columns={
    'Type':'type',
    'Per Serving': 'per_serving',
    'Total Fat': 'total_fat',
    'Sodium': 'sodium',
    'Sugars': 'sugars'},inplace=True)
cracker.head()
```

```python
cream.head()
```

```python
cream = cream[['type','product','Per Serving','Total Fat', 'Sodium', 'Sugars']]
cream.rename(columns={
    'Per Serving': 'per_serving',
    'Total Fat': 'total_fat',
    'Sodium': 'sodium',
    'Sugars': 'sugars'},inplace=True)
cream.head()
```

```python
wafer.head()
```

```python
wafer = wafer[['type','product','Per Serving','Total Fat', 'Sodium', 'Sugars']]
wafer.rename(columns={
    'Per Serving': 'per_serving',
    'Total Fat': 'total_fat',
    'Sodium': 'sodium',
    'Sugars': 'sugars'},inplace=True)
wafer.head()
```

```python
df = pd.concat([cookie,cracker,cream,wafer])
df.reset_index(inplace=True)
df.drop('index',axis =1, inplace=True)
df.head(100)
```

```python
# Remove 'g' from the 'total_fat' and 'sugars' columns
df['per_serving_g'] = df['per_serving'].str.replace('g', '').astype(float)
df['total_fat_g'] = df['total_fat'].str.replace('g', '').astype(float)
df['sugars_g'] = df['sugars'].str.replace('g', '').astype(float)
df['sodium_mg'] = df['sodium'].str.replace('mg', '').astype(float)

df.head(100)
```

```python
df_clean = df.drop(['per_serving','total_fat','sodium','sugars'], axis=1)
df_clean.head()
```

```python
df_clean.shape
```

```python
df_clean.dtypes
```

### Divide nutrients by per serving size

```python
df_clean['total_fat_g_per_gram_of_serving'] = df_clean['total_fat_g'] / df_clean['per_serving_g']
df_clean['sugars_g_per_gram_of_serving'] = df_clean['sugars_g'] / df_clean['per_serving_g']
df_clean['sodium_mg_per_gram_of_serving'] = df_clean['sodium_mg'] / df_clean['per_serving_g']
```

```python
df_clean.head(100)
```

```python
df_final = df_clean.drop(['per_serving_g','total_fat_g','sugars_g','sodium_mg'], axis=1)
df_final.head(100)
```

### EDA

```python
# Set the figure size
plt.figure(figsize=(6,4))

# Create the histogram
sns.histplot(data=df_final,x="total_fat_g_per_gram_of_serving", bins=100)

# Add labels and titles
plt.xlabel('Total Fat per 1g of serving')
plt.ylabel('Count')
plt.title("Distribution of Total Fat")

# Show the plot
plt.show()
```

```python
# Set the style of the plot
sns.set(style="whitegrid")

# Set the size of the plot
plt.figure(figsize=(10, 4))

# Plot separate boxplots for each type
sns.boxplot(data=df_final, x='total_fat_g_per_gram_of_serving', y='type', orient='h')

# Set titles and labels of the plot
plt.title('Boxplot of Total Fat per 1g of serving')
plt.ylabel('Type')
plt.xlabel('Total Fat')

# Show the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(6,4))

# Create the histogram
sns.histplot(data=df_final,x="sugars_g_per_gram_of_serving", bins=100)

# Add labels and titles
plt.xlabel('Sugars per 1g of serving')
plt.ylabel('Count')
plt.title("Distribution of Sugars")

# Show the plot
plt.show()
```

```python
# Set the style of the plot
sns.set(style="whitegrid")

# Set the size of the plot
plt.figure(figsize=(10, 4))

# Plot separate boxplots for each type
sns.boxplot(data=df_final, x='sugars_g_per_gram_of_serving', y='type', orient='h')

# Set titles and labels of the plot
plt.title('Boxplot of Sugars per 1g of serving')
plt.ylabel('Type')
plt.xlabel('Sugar')

# Show the plot
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(6,4))

# Create the histogram
sns.histplot(data=df_final,x="sodium_mg_per_gram_of_serving", bins=100)

# Add labels and titles
plt.xlabel('Sodium per 1g of serving')
plt.ylabel('Count')
plt.title("Distribution of Sodium")

# Show the plot
plt.show()
```
